```python
import os, json, warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from dotenv import load_dotenv

load_dotenv()
warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
os.makedirs("charts", exist_ok=True)

API_URL = os.getenv('API_URL', 'https://data-charts-api.hexlet.app')
DATE_BEGIN = os.getenv('DATE_BEGIN', '2023-03-01')
DATE_END = os.getenv('DATE_END', '2023-09-01')

# Загрузка данных
try:
    rv = requests.get(f"{API_URL}/visits", params={'begin': DATE_BEGIN, 'end': DATE_END}, verify=False, timeout=30)
    rr = requests.get(f"{API_URL}/registrations", params={'begin': DATE_BEGIN, 'end': DATE_END}, verify=False, timeout=30)
    rv.raise_for_status(); rr.raise_for_status()
    df_visits_api = pd.DataFrame(rv.json())
    df_reg_api = pd.DataFrame(rr.json())
except Exception as e:
    print(f"API Error: {e}"); df_visits_api, df_reg_api = pd.DataFrame(), pd.DataFrame()

try:
    df_ads = pd.read_csv('./ads.csv')
except:
    df_ads = pd.DataFrame()

print(f"✅ Данные: {len(df_visits_api)} визитов, {len(df_reg_api)} рег., {len(df_ads)} рекламы")
```

```python
if not df_visits_api.empty:
    # --- Предобработка визитов ---
    df_v = df_visits_api.copy()
    df_v['datetime'] = pd.to_datetime(df_v['datetime'])
    df_v = df_v[~df_v['user_agent'].str.contains('bot', case=False, na=False)]
    df_v = df_v.sort_values('datetime').drop_duplicates(subset=['visit_id'], keep='last')
    df_v['date_group'] = df_v['datetime'].dt.date

    # --- Предобработка регистраций ---
    df_r = df_reg_api.copy()
    df_r['datetime'] = pd.to_datetime(df_r['datetime'])
    df_r['date_group'] = df_r['datetime'].dt.date

    # --- 1. Агрегация данных ---
    v_agg = df_v.groupby(['date_group', 'platform']).size().reset_index(name='visits')
    r_agg = df_r.groupby(['date_group', 'platform']).size().reset_index(name='registrations')
    
    df_conv = pd.merge(v_agg, r_agg, on=['date_group', 'platform'], how='outer').fillna(0)
    df_conv['visits'] = df_conv['visits'].astype(int)
    df_conv['registrations'] = df_conv['registrations'].astype(int)
    df_conv['conversion'] = (df_conv['registrations'] / df_conv['visits'] * 100).round(2)
    
    # Сортировка
    df_conv['platform'] = pd.Categorical(df_conv['platform'], categories=['android', 'ios', 'web'], ordered=True)
    df_conv = df_conv.sort_values(['date_group', 'platform']).reset_index(drop=True)
    df_conv['platform'] = df_conv['platform'].astype(str)
    
    # 🔧 СОХРАНЕНИЕ В ПРАВИЛЬНОМ ФОРМАТЕ (чистые списки, без индексных ключей)
    df_conv_export = df_conv.copy()

df_conv_export['date_group'] = (
    pd.to_datetime(df_conv_export['date_group']).astype(np.int64) // 10**6
)

df_conv_export.to_json('./conversion.json')
    print("✅ conversion.json сохранён")

    # --- 2. Сохранение ads.json ---
    if not df_ads.empty:
        v_d = df_v.groupby('date_group').size().reset_index(name='visits')
        r_d = df_r.groupby('date_group').size().reset_index(name='registrations')
        df_ads['date_group'] = pd.to_datetime(df_ads['date']).dt.date
        ads_agg = df_ads.groupby('date_group').agg({'cost': 'sum', 'utm_campaign': 'first'}).reset_index()
        
        df_final = pd.merge(v_d, r_d, on='date_group', how='outer').fillna(0)
        df_final = pd.merge(df_final, ads_agg, on='date_group', how='left')
        df_final['visits'] = df_final['visits'].astype(int)
        df_final['registrations'] = df_final['registrations'].astype(int)
        df_final['cost'] = df_final['cost'].fillna(0).astype(int)
        df_final['utm_campaign'] = df_final['utm_campaign'].fillna('none')
        df_final = df_final.sort_values('date_group').reset_index(drop=True)
        df_final[['date_group', 'visits', 'registrations', 'cost', 'utm_campaign']].to_json('./ads.json')
        print("✅ ads.json сохранён")

    # --- 3. Подготовка переменных для графиков ---
    visits_total = df_v.groupby('date_group').size().reset_index(name='visits')
    regs_total = df_r.groupby('date_group').size().reset_index(name='registrations')
    visits_by_date_platform = df_v.groupby(['date_group', 'platform']).size().unstack(fill_value=0).reset_index()
    reg_by_date_platform = df_r.groupby(['date_group', 'platform']).size().unstack(fill_value=0).reset_index()
    df_conv_plat = pd.merge(v_agg, r_agg, on=['date_group', 'platform'], how='outer').fillna(0)
    df_conv_plat['conversion'] = (df_conv_plat['registrations'] / df_conv_plat['visits'] * 100).round(0)

    print("✅ Данные для графиков готовы")
```

```python
colors = {'android': '#4A90E2', 'ios': '#F5A623', 'web': '#50E3C2'}

# 1. Итоговые визиты
plt.figure(figsize=(14, 6), dpi=100)
bars = plt.bar(visits_total['date_group'], visits_total['visits'], color='skyblue', edgecolor='black', linewidth=0.5)
for bar, val in zip(bars, visits_total['visits']):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 15, f'{int(val)}', ha='center', va='bottom', fontsize=9, fontweight='bold')
plt.title('Total Visits', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('date_group', fontsize=11); plt.ylabel('visits', fontsize=11)
plt.xticks(rotation=45, ha='right', fontsize=8); plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('charts/total_visits.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/total_visits.png")

# 2. Визиты по платформам
plt.figure(figsize=(14, 6), dpi=100)
plt.bar(visits_by_date_platform['date_group'], visits_by_date_platform.get('android', 0), color=colors['android'], label='android', edgecolor='white', linewidth=0.5)
plt.bar(visits_by_date_platform['date_group'], visits_by_date_platform.get('ios', 0), bottom=visits_by_date_platform.get('android', 0), color=colors['ios'], label='ios', edgecolor='white', linewidth=0.5)
# 🔧 ИСПРАВЛЕНО: color=colors['web'] (убран пробел в "col or")
plt.bar(visits_by_date_platform['date_group'], visits_by_date_platform.get('web', 0), bottom=visits_by_date_platform.get('android', 0) + visits_by_date_platform.get('ios', 0), color=colors['web'], label='web', edgecolor='white', linewidth=0.5)
plt.title('Visits by Platform (Stacked)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('date_group', fontsize=11); plt.ylabel('visits', fontsize=11)
plt.xticks(rotation=45, ha='right', fontsize=8); plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='platform', loc='upper right'); plt.tight_layout()
plt.savefig('charts/visits_by_platform.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/visits_by_platform.png")

# 3. Итоговые регистрации
plt.figure(figsize=(14, 6), dpi=100)
bars = plt.bar(regs_total['date_group'], regs_total['registrations'], color='skyblue', edgecolor='black', linewidth=0.5)
for bar, val in zip(bars, regs_total['registrations']):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 5, f'{int(val)}', ha='center', va='bottom', fontsize=9, fontweight='bold')
plt.title('Total Registrations', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('date_group', fontsize=11); plt.ylabel('registrations', fontsize=11)
plt.xticks(rotation=45, ha='right', fontsize=8); plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('charts/total_registrations.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/total_registrations.png")

# 4. Регистрации по платформам
plt.figure(figsize=(14, 6), dpi=100)
plt.bar(reg_by_date_platform['date_group'], reg_by_date_platform.get('android', 0), color=colors['android'], label='android', edgecolor='white', linewidth=0.5)
plt.bar(reg_by_date_platform['date_group'], reg_by_date_platform.get('ios', 0), bottom=reg_by_date_platform.get('android', 0), color=colors['ios'], label='ios', edgecolor='white', linewidth=0.5)
# 🔧 ИСПРАВЛЕНО: colors['web'] (убран пробел в 'w eb')
plt.bar(reg_by_date_platform['date_group'], reg_by_date_platform.get('web', 0), bottom=reg_by_date_platform.get('android', 0) + reg_by_date_platform.get('ios', 0), color=colors['web'], label='web', edgecolor='white', linewidth=0.5)
plt.title('Registrations by Platform (Stacked)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('date_group', fontsize=11); plt.ylabel('registrations', fontsize=11)
plt.xticks(rotation=45, ha='right', fontsize=8); plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='platform', loc='upper right'); plt.tight_layout()
plt.savefig('charts/registrations_by_platform.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/registrations_by_platform.png")

# 5. Средняя конверсия
df_conv_trend = pd.merge(visits_total, regs_total, on='date_group')
df_conv_trend['conversion'] = (df_conv_trend['registrations'] / df_conv_trend['visits'] * 100).round(0)
plt.figure(figsize=(12, 6), dpi=100)
plt.plot(df_conv_trend['date_group'], df_conv_trend['conversion'], color='#4A90E2', linewidth=1.5, marker='o', markersize=6, label='Общая конверсия')
for i, row in df_conv_trend.iterrows():
    offset = 1.5 if i % 2 == 0 else -2
    plt.text(row['date_group'], row['conversion'] + offset, f"{int(row['conversion'])}%", ha='center', va='bottom' if offset > 0 else 'top', fontsize=9, color='black')
plt.title('Average Conversion', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Date', fontsize=11); plt.ylabel('Conversion (%)', fontsize=11)
plt.xticks(rotation=45, ha='right', fontsize=9); plt.grid(True, linestyle='-', alpha=0.4)
plt.legend(loc='upper right'); plt.tight_layout()
plt.savefig('charts/average_conversion.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/average_conversion.png")

# 6. Конверсия по каждой платформе
fig, axes = plt.subplots(3, 1, figsize=(12, 14), dpi=100, sharex=True)
platforms = ['android', 'ios', 'web']
for ax, platform in zip(axes, platforms):
    data = df_conv_plat[df_conv_plat['platform'] == platform].sort_values('date_group')
    if data.empty: continue
    ax.plot(data['date_group'], data['conversion'], color=colors[platform], linewidth=1.5, marker='o', markersize=4, label=platform)
    for i, row in data.iterrows():
        ax.text(row['date_group'], row['conversion'] + 2, f"{int(row['conversion'])}%", ha='center', fontsize=8, color='black')
    ax.set_title(f'Conversion {platform}', fontsize=12, fontweight='bold', pad=10)
    ax.set_ylabel('Conversion (%)', fontsize=10); ax.grid(True, linestyle='--', alpha=0.5)
    ax.set_ylim(0, data['conversion'].max() * 1.2 + 10); ax.legend(loc='upper left')
    if platform == 'web': ax.tick_params(axis='x', rotation=45)
plt.xlabel('Date', fontsize=11); plt.tight_layout()
plt.savefig('charts/conversion_by_platform.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/conversion_by_platform.png")

# 7. Стоимости реклам
ads_cost = df_ads.groupby('date_group')['cost'].sum().reset_index()
plt.figure(figsize=(14, 6), dpi=100)
plt.plot(ads_cost['date_group'], ads_cost['cost'], color='#4A90E2', linewidth=1.5, marker='o', markersize=5)
for i, row in ads_cost.iterrows():
    offset = 12 if i % 2 == 0 else -25
    plt.text(row['date_group'], row['cost'] + offset, f"{int(row['cost'])} RUB", ha='center', va='bottom' if offset > 0 else 'top', fontsize=8, color='black', fontweight='bold')
plt.title('Aggregated Ad Campaign Costs (by day)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Date', fontsize=11); plt.ylabel('Cost (RUB)', fontsize=11)
plt.xticks(rotation=45, ha='right', fontsize=8); plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('charts/ad_costs.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/ad_costs.png")

# 8. Визиты с выделением рекламных кампаний
import matplotlib.colors as mcolors
v_ads = visits_total.copy()
v_ads['date_group'] = pd.to_datetime(v_ads['date_group'])
df_ads_plot = df_ads.copy()
df_ads_plot['date_group'] = pd.to_datetime(df_ads_plot['date_group'])
fig, ax = plt.subplots(figsize=(14, 6))
avg_visits = v_ads['visits'].mean()
ax.plot(v_ads['date_group'], v_ads['visits'], color='black', marker='o', markersize=5, linewidth=1.5, label='Visits')
ax.axhline(y=avg_visits, color='gray', linestyle='--', linewidth=1, label='Average Number of Visits')
campaign_groups = df_ads_plot.groupby(['utm_source', 'utm_medium', 'utm_campaign'])
tbl_colors = list(mcolors.TABLEAU_COLORS.values())
for i, ((src, med, camp), group) in enumerate(campaign_groups):
    start_date = group['date_group'].min()
    end_date = group['date_group'].max()
    label_name = f"{src} {med} {camp}"
    ax.axvspan(start_date, end_date, color=tbl_colors[i % len(tbl_colors)], alpha=0.4, label=label_name)
ax.set_title("Visits during marketing active days", fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel("Unique Visits", fontsize=12); ax.set_xlabel("Date", fontsize=12)
ax.legend(loc='lower left', fontsize=10); ax.grid(True, linestyle='--', alpha=0.3)
fig.autofmt_xdate(rotation=45); plt.tight_layout()
plt.savefig('charts/visits_with_ads.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/visits_with_ads.png")

# 9. Регистрации с выделением рекламных кампаний
r_ads = regs_total.copy()
r_ads['date_group'] = pd.to_datetime(r_ads['date_group'])
fig, ax = plt.subplots(figsize=(14, 6))
avg_regs = r_ads['registrations'].mean()
ax.plot(r_ads['date_group'], r_ads['registrations'], color='#2E8B57', marker='o', markersize=5, linewidth=1.5, label='Registrations')
ax.axhline(y=avg_regs, color='gray', linestyle='--', linewidth=1, label='Average Number of Registration')
tbl_colors = list(mcolors.TABLEAU_COLORS.values())
for i, ((src, med, camp), group) in enumerate(campaign_groups):
    start_date = group['date_group'].min()
    end_date = group['date_group'].max()
    label_name = f"{src} {med} {camp}"
    ax.axvspan(start_date, end_date, color=tbl_colors[i % len(tbl_colors)], alpha=0.4, label=label_name)
ax.set_title("Registrations during marketing active days", fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel("Unique Users", fontsize=12); ax.set_xlabel("Date", fontsize=12)
ax.legend(loc='upper right', fontsize=10); ax.grid(True, linestyle='-', alpha=0.3)
fig.autofmt_xdate(rotation=45); plt.tight_layout()
plt.savefig('charts/registrations_with_ads.png', dpi=150, bbox_inches='tight'); plt.close()
print("✅ charts/registrations_with_ads.png")
```
